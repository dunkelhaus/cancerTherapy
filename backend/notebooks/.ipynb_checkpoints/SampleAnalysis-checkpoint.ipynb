```python
multidiffs = [()]

def ifNA(value):
    if type(value) == "float":
        return False
    if value[0] == 'N':
        return True
    else:
        return False
    
    return

for i in range(1, len(toprow.columns) - 2):
    mel1 = melanoma1.columns[i]
    mel10 = melanoma10.columns[i]
    lung1 = lungad1.columns[i]
    lung10 = lungad10.columns[i]
    if ifNA(mel1) == True:
        mel1 = 0.0
    if ifNA(mel10) == True:
        mel10 = 0.0
    if ifNA(lung1) == True:
        lung1 = 0.0
    if ifNA(lung10) == True:
        lung10 = 0.0
    if ifNA(mel1) == False:
        mel1 = float(melanoma1.columns[i])
    if ifNA(mel10) == False:
        mel10 = float(melanoma10.columns[i])
    if ifNA(lung1) == False:
        lung1 = float(lungad1.columns[i])
    if ifNA(lung10) == False:
        lung10 = float(lungad10.columns[i])
    multidiffs.append((mel1, lung1, mel10, lung10))
```
