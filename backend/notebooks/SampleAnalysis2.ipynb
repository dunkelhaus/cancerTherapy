```python
multidiffs = [()]
f = open("/home/skjena/cancerTherapy/backend/out/occurence.csv", "w+")
f.write("Sequence,Melanoma[1],Melanoma[10],LungAdeno[1],LungAdeno[10]\n")

def ifNA(value):
    if type(value) == str:
        if value[0] == 'N':
            return -1
        else:
            return int(value)
    return

def occurence(value):
    splitlist = value.split('.')
    return splitlist[0]

for i in range(1, len(toprow.columns) - 2):
    seq = toprow.columns[i]
    mel1 = ifNA(occurence(melanoma1.columns[i]))
    mel10 = ifNA(occurence(melanoma10.columns[i]))
    lung1 = ifNA(occurence(lungad1.columns[i]))
    lung10 = ifNA(occurence(lungad10.columns[i]))
    f.write("%s,%s,%s,%s,%s\n" %(seq, mel1, mel10, lung1, lung10))
    multidiffs.append((seq, mel1, mel10, lung1, lung10))
```

```python
recurlist = [()]
f = open("/home/skjena/cancerTherapy/backend/out/recurrence.csv", "w+")
f.write("Sequence,Melanoma[1],Melanoma[1]recurrence,Melanoma[10],Melanoma[10]recurrence,LungAdeno[1],LungAdeno[1]recurrence,LungAdeno[10],LungAdeno[10]recurrence\n")

def ifNA(value):
    if value[0] == 'N':
        return -1
    else:
        return int(value)
    return

def recurrence(value):
    splitlist = value.split('.')
    
    if len(splitlist) > 1:
        retval = int(splitlist[1])
    else:
        retval = 0
    return retval

for i in range(1, len(toprow.columns) - 2):
    seq = toprow.columns[i]
    mel1 = ifNA(occurence(melanoma1.columns[i]))
    mel10 = ifNA(occurence(melanoma10.columns[i]))
    lung1 = ifNA(occurence(lungad1.columns[i]))
    lung10 = ifNA(occurence(lungad10.columns[i]))
    mel1rec = recurrence(melanoma1.columns[i])
    mel10rec = recurrence(melanoma10.columns[i])
    lung1rec = recurrence(lungad1.columns[i])
    lung10rec = recurrence(lungad10.columns[i])
    f.write("%s,%s,%s,%s,%s,%s,%s,%s,%s\n" %(seq, mel1, mel1rec, mel10, mel10rec, lung1, lung1rec, lung10, lung10rec))
    recurlist.append((seq, mel1, mel1rec, mel10, mel10rec, lung1, lung1rec, lung10, lung10rec))
```
